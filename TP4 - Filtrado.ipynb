c) Diseñe **al menos** dos filtros FIR y dos IIR para su comparación. Verifique que la respuesta en frecuencia responda a la plantilla de diseño
  

```python
## Desarrollo de filtros IIR.

## Frecuencia de muestreo.
fs = 1000

## Plantilla de diseño.
fs0 = 1 # fin de la banda de detenida 0.
fc0 = 2 # comienzo de la banda de paso.
fc1 = 25 # fin de la banda de paso.
fs1 = 50 # comienzo de la banda de detenida 1..
gpass = 1 # Ganancia en banda de paso
gstop = 40 # Atenuación banda de stop.

## Normalizo las frecuencia de las bandas de paso y stop.
wp = npy.array([fc0,fc1]) / (fs/2)
ws = npy.array([fs0,fs1]) / (fs/2)


## Diseño de filtro iir tipo Butterworth.
butter = sig.iirdesign(wp,ws, gpass, gstop,analog=False,output='sos',ftype='butter')
w1, h1 = sig.sosfreqz(butter)

## Diseño de filtro iir tipo Chebyshev II.
cheby = sig.iirdesign(wp,ws,gpass, gstop,ftype='cheby1',analog=False,output='sos')
w2, h2 = sig.sosfreqz(cheby)

## Ploteo.
plt.figure(1,figsize=(20,8))
plt.title('Filtro IIR Butterworth')
plt.xlim(0,60)
plt.ylim(-50,1)
plt.plot(w1*(fs/2)/npy.pi, 20 * npy.log10(abs(h1)),label='butter')
plt.ylabel('Amplitud [dB]')
plt.xlabel('Frequencia [Hz]')
plt.grid()
plt.legend()

plt.figure(2,figsize=(20,8))
plt.title('Filtro IIR Chebyshev I')
plt.xlim(0,100)
plt.ylim(-60,1)
plt.plot(w2*(fs/2)/npy.pi, 20 * npy.log10(abs(h2)),label='chebyshev1')
plt.ylabel('Amplitud [dB]')
plt.xlabel('Frequencia [Hz]')
plt.grid()
plt.legend()
```

```python
## Desarrollo de filtros FIR.

## Frecuencia de muestreo.
fs = 1000

## Plantilla de diseño.
fs0 = 1 # fin de la banda de detenida 0.
fc0 = 2 # comienzo de la banda de paso.
fc1 = 25 # fin de la banda de paso.
fs1 = 50 # comienzo de la banda de detenida 1..
gpass = 1 # Ganancia en banda de paso
gstop = 40 # Atenuación banda de stop.

coeficientes = 501

# Array de frecuencias normalizadas de 0 a 1 (nyquist)
frecs = npy.array([0,fs0,fc0,fc1,fs1,fs/2])

# Array de ganancia.
gan = npy.array([gstop,gstop,gpass,gpass,gstop,0]) * (-1)
gan = 10**(gan/20) # Paso a veces.
gan[5] = 0 # En fs/2 debe ser 0 la ganancia.

gan1 = npy.array([gstop,gstop,gpass,gpass,gstop,0]) * (-1)
gan1 = 10**(gan1/20) # Paso a veces.

# Diseño de filtro FIR por método de ventana.
fwin = sig.firwin2(coeficientes,frecs,gan,fs=fs,window='boxcar')
wa, ha = sig.freqz(fwin,1)

# Diseño de filtro FIR por método de los cuadrados menos significativos.
#fls = sig.firls(coeficientes, frecs, gan1, fs=fs)
fls = sig.firls(61, frecs, [0.001,0.001,1,1,0.001,0.001], fs=fs)

# Tuve que modificar el orden y los valores en la atenuación porque no reflejaba los valores de la plantilla.

wb,hb = sig.freqz(fls,1)

#Ploteo.
plt.figure(1,figsize=(20,8))
plt.title('Filtro FIR con método de ventana.')
plt.xlim(0,70)
plt.ylim(-55,1)
plt.plot((fs/2)*wa/npy.pi,20*npy.log10(abs(ha)))
plt.ylabel('Amplitud [dB]')
plt.xlabel('Frequencia [Hz]')
plt.grid()

plt.figure(2,figsize=(20,8))
plt.title('Filtro FIR con método less-square.')
plt.xlim(0,100)
plt.ylim(-55,1)
plt.plot((fs/2)*wb/npy.pi,20*npy.log10(abs(hb)))
plt.ylabel('Amplitud [dB]')
plt.xlabel('Frequencia [Hz]')
plt.grid()
```
